```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .6)
```

```python
font = {'family': 'DIN Alternate',
        'weight': 'bold',
        'size': 14,
        }
matplotlib.rc('font', **font)

xgboost_labels = ['Density', 'Geographical Risk', 'Number of Housing Units','Total Population', 'Bachelors Rate', 'Area of Zipcode',]
ann_labels = ['Number of Housing Units', 'Total Population', 'Density', 'Percentage Black', 'Bachelors Rate', 'Geographical Risk']
L = [xgboost_labels, ann_labels]
export_labels = ['xgboost', 'ann']


xgboost_imps =[.2185, .079, .065, .094, .084, .080]
ann_imps = [.2054, .2016, .1984, .1977, .1930, .1868,]
res = [xgboost_imps, ann_imps]

for i in range(2):
    zipped = sorted(zip(L[i], res[i]), key = lambda x: x[-1], reverse = True)

    dic = {e[0]: e[1] for e in zipped}

    plt.figure()
    plt.bar([x for x in range(len(res[i]))], dic.values())

    fig, ax = plt.subplots(1,1, figsize=(8, 5))
    ax = sns.barplot(x=list(range(6)), y=[e[1] for e in zipped], )
    ax.set(xlabel= 'Features', ylabel='Importance', xticklabels=[e for e in dic])
    #ax.legend(loc=6, bbox_to_anchor=(0.55, 0.75), labels=dic)
    plt.setp(ax.get_xticklabels(), rotation=20, ha="right", rotation_mode="anchor")

    change_width(ax, .6)
    fig.savefig(f'{export_labels[i]}_feature_importances.png', dpi=330, bbox_inches='tight')
```
